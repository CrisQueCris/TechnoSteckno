```python
## Building a dataframe of spotify top 100 last 10 weeks track id and popularity

result={'artist': None,'id':None,'title':None,'popularity':None}
spot_pop = pd.DataFrame(columns=result)
    
```

```python
## Only Get track id:

for title in list(billboard10['title']):
    print(title)
    if billboard10.loc[billboard10['title']==title, 'id'].index[0] == 0:
        result_search = sp.search(q=title[0:10], limit=1, market='GB')
        result['id']= result_search["tracks"]["items"][0]["id"] # Track ID
        print(result['id'])
        billboard10.loc[billboard10['title']==title, 'id']=result['id']
        seconds_to_wait = randint(1,3)
        seconds_to_wait
        sleep(seconds_to_wait)
```
